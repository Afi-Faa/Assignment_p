```python
import pandas as pd
import numpy as np
import warnings as wr
wr.filterwarnings('ignore')
```

```python
x = np.array([21,20,22,22,23,21,25,30,31])
y = np.array([60,55,60,61,65,62,65,70,68])
```

```python
df = pd.DataFrame({'age': x, 'Income(k)': y})
df
```

```python
target_age = 22
df['dist'] = abs(df['age'] - target_age)
df_sorted = df.sort_values(by='dist')
k = 3
nearest_neighbors = df_sorted.iloc[:k] #iloc = integer based indexing [kind of slicing technique]
predicted_inc = nearest_neighbors['Income(k)'].mean()
print(f"Predicted income for age in KNN {target_age} is {predicted_inc:.2f}k.")
```

```python
df.shape
```

## Linear Regreession

```python
from sklearn.linear_model import LinearRegression
```

```python
df = df.drop('dist', axis = 1) ## dont need for Linear Regression 
```

```python
X = df[['age']]
Y = df['Income(k)']
```

```python
model = LinearRegression()
model.fit(X,Y)
```

```python
predicted_income = model.predict([[target_age]])[0] # [] extracts the scaler value bcz it returns numpy array
```

```python
predicted_income
```

```python
print(f"Predicted income for age in Linear regression {target_age} is {predicted_income:.2f}k.")
```

## Comparison

```python
import matplotlib.pyplot as plt
m = ['KNN', 'Linear Regression']
predictions = [predicted_inc, predicted_income]
plt.bar(m, predictions, color=['blue', 'green'])
plt.ylabel('Predicted Income(k)')
plt.title('KNN vs Linear Regression Prediction for Age 22')
plt.show()
```
